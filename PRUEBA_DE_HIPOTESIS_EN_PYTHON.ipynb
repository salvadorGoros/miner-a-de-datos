Problema 1: Control de Calidad en una Fábrica de Tornillos
Una fábrica de tornillos afirma que el diámetro promedio de sus tornillos es de 10 mm.
Se toma una muestra aleatoria de 49 tornillos, obteniéndose un diámetro promedio de 9.7 mm con una desviación estándar de 0.5 mm.

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# Datos
nu_0 = 10
n = 49
x_barra = 9.7
s = 0.5
alpha = 0.01

# Estadístico t
t = (x_barra - nu_0) / (s / np.sqrt(n))

# Valor crítico (cola izquierda)
t_critical = stats.t.ppf(alpha, df=n-1)

# Decisión
if t < t_critical:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")

print("t calculado:", t)
print("t crítico:", t_critical)

df = pd.DataFrame({
    'Parametros':['Media Teorica','Media Muestral','Desviación Estándar','n','alpha'],
    'Valores':[nu_0, x_barra, s, n, alpha]
})

print(df)
```

¿Se puede concluir, con un nivel de significancia del 1%, que el diámetro medio es menor al teórico de 10 mm?

R=Sí, el diámetro es significativamente menor a 10 mm. La diferencia de 0.3 mm es estadísticamente importante para un nivel del 1%.

Problema 2: Duración de Baterías de Celulares
Un fabricante de celulares asegura que la duración promedio de sus baterías es de 20 horas. Se toma una muestra de 30 baterías y se encuentra que su duración promedio es de 18.5 horas, con una desviación estándar de 2.5 horas.  

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# Datos del Problema 2
mu_0 = 20         # Media teórica
n = 30            # Tamaño de muestra
x_barra = 18.5    # Media muestral
s = 2.5           # Desviación estándar
alpha = 0.05      # Nivel de significancia

# Estadístico t
t_stat = (x_barra - mu_0) / (s / np.sqrt(n))

# Valor crítico (cola izquierda)
t_critical = stats.t.ppf(alpha, df=n-1)

# Decisión
decision = "Rechazamos la hipótesis nula" if t_stat < t_critical else "No rechazamos la hipótesis nula"

print(f"Resultado: {decision}")
print(f"t calculado: {t_stat:.4f}")
print(f"t crítico: {t_critical:.4f}")

# Mostrar en DataFrame
df = pd.DataFrame({
    'Parámetros': ['Media Teórica', 'Media Muestral', 'Desviación Estándar', 'n', 'alpha'],
    'Valores': [mu_0, x_barra, s, n, alpha]
})
print("\nResumen de datos:")
print(df)
```

Con un nivel de significancia del 5%, ¿se puede concluir que la duración de las baterías es menor a la especificada por el fabricante?  

R=Sí, la duración es menor a las 20 horas prometidas. El promedio de 18.5 horas cae fuera del margen de error aceptable al 5%.

Problema 3: Control de Peso en una Planta de Alimentos
Un productor de harina en bolsa indica que el peso promedio de sus bolsas es de 1 kg (1000 gramos). Para verificarlo, se toman 40 bolsas y se obtiene un peso promedio de 990 gramos, con una desviación estándar de 12 gramos.  

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# Datos
nu_0 = 1000
n = 40
x_barra = 990
s = 12
alpha = 0.02

# Estadístico t
t = (x_barra - nu_0) / (s / np.sqrt(n))

# Valor crítico (cola izquierda)
t_critical = stats.t.ppf(alpha, df=n-1)

# Decisión
if t < t_critical:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")

print("t calculado:", t)
print("t crítico:", t_critical)

df = pd.DataFrame({
    'Parametros':['Media Teorica','Media Muestral','Desviación Estándar','n','alpha'],
    'Valores':[nu_0, x_barra, s, n, alpha]
})

print(df)
```

Si se usa un nivel de significancia del 2%, ¿se puede concluir que las bolsas tienen un peso menor al anunciado?

R=Sí, las bolsas tienen un peso menor al anunciado. Con un 2% de significancia, los 10 gramos de menos no se deben al azar.
