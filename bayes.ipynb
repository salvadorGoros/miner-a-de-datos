```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import numpy as np
import pandas as pd

data={
    'correo': [
        "gana dinero rapido",
        "oferta exclusiva gratis",
        "reunion mañana",
        "adjunto el reporte",
        "ganapremio gratis"
    ],
    'clase': [
        "spam",
        "spam",
        "no spam",
        "no spam",
        "spam",
    ]
}

df= pd.DataFrame(data)
print(df)
```

```python
# Convertir el texto en una matriz de características
vectorizador = CountVectorizer()
x= vectorizador.fit_transform(df['correo'])
print(x.toarray())
```

```python
# Variable objetivo
y = df['clase']
# Entrenar el modelo
modelo = MultinomialNB()
modelo.fit(x,y)
# Predecir la clase de un nuevo correo
nuevo_correo = ['reunion ayer']
correo_vectorizado = vectorizador.transform(nuevo_correo)
prediccion = modelo.predict(correo_vectorizado)
print(prediccion)
```

```python
palabras = vectorizador.get_feature_names_out()
tabla_prob = pd.DataFrame(
    modelo.feature_log_prob_,
    columns=palabras,
    index=modelo.classes_
)
print(tabla_prob)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import numpy as np
import pandas as pd

data ={
    "reseña":[
          "el producto funciona excelente",
        "muy buena calidad",
        "el producto llegó dañado",
        "muy mala calidad",
        "excelente producto y calidad"
    ],
    "clasificacion":[
            "satisfecho",
        "satisfecho",
        "insatisfecho",
        "insatisfecho",
        "satisfecho"
    ]

}
df = pd.DataFrame(data)
```

```python
#vectorizacion del texto
# Re-create df and y for sentiment analysis to ensure correct data is used
data_sentiment = {
    "reseña":[
          "el producto funciona excelente",
        "muy buena calidad",
        "el producto llegó dañado",
        "muy mala calidad",
        "excelente producto y calidad"
    ],
    "clasificacion":[
            "satisfecho",
        "satisfecho",
        "insatisfecho",
        "insatisfecho",
        "satisfecho"
    ]
}
df_sentiment = pd.DataFrame(data_sentiment)

vectorizer = CountVectorizer()
x = vectorizer.fit_transform(df_sentiment['reseña'])
print(x.toarray())

# Define the target variable 'y' from the sentiment analysis DataFrame
y = df_sentiment['clasificacion']

#entrenar modelo
modelo = MultinomialNB() #sepacion de datos
modelo.fit(x,y) # entrena por separado
# Comentario nuevo
nuevo = ["producto mala calidad"]
X_new = vectorizer.transform(nuevo)

# Predicción
prediccion = modelo.predict(X_new) # Corrected 'model' to 'modelo'
probabilidades = modelo.predict_proba(X_new) # Corrected 'model' to 'modelo'

print("Comentario:", nuevo[0])
print("Clasificación predicha:", prediccion[0])
print("Probabilidades:", probabilidades)
```
