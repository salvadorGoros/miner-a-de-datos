```python
#ejercicico 1
import numpy as np
from scipy import stats

datos = np.array([498, 501, 499, 502, 500, 497, 503, 499, 501, 500])

t_stat, p_value = stats.ttest_1samp(datos, popmean=500)

print("Estadístico t:", t_stat)
print("p-value:", p_value)
```

```python
#ejercicio 2
import numpy as np
from scipy import stats

grupo_musica = np.array([65, 70, 68, 72, 66, 69, 71, 67, 70, 68])
grupo_silencio = np.array([85, 88, 90, 87, 92, 86, 89, 91, 88, 90])

t_stat, p_value = stats.ttest_ind(grupo_musica, grupo_silencio, equal_var=False)

print("Estadístico t:", t_stat)
print("p-value:", p_value)
```

```python
#ejercicio 3
import numpy as np
from scipy import stats

# Datos observados
observados = np.array([200, 120, 80])

n = np.sum(observados)

proporciones = np.array([0.40, 0.35, 0.25])

esperados = proporciones * n

chi2, p_value = stats.chisquare(f_obs=observados, f_exp=esperados)

print("Chi-cuadrado:", chi2)
print("p-value:", p_value)

gl = len(observados) - 1
print("Grados de libertad:", gl)
```
