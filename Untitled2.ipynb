```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
data = {
    "Horas de estudio": [1,2,1.5,8,9,8.5,3,3.5,4],
    "Calificaciones": [2,3,2.5,9,10,9.5,5,5.5,6]
}
df=pd.DataFrame(data)
df
```

```python
plt.scatter(df["Horas de estudio"], df["Calificaciones"])
plt.xlabel("Horas de estudio")
plt.ylabel("Calificaciones")
plt.title("Datos de los estudiantes")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(df)
```

```python
centroides = kmeans.cluster_centers_
centroides
```

```python
df['Cluster'] = kmeans.labels_
df
```

```python
plt.scatter(df["Horas de estudio"], df["Calificaciones"],c=df['Cluster'])
plt.scatter(centroides[:,0], centroides[:,1], marker="x",color="red",s=200, label="Centroides")
```

```python
# 🔹 MÉTODO DEL CODO
inercia = []

for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df)
    inercia.append(kmeans.inertia_)

# 🔹 Gráfica del codo
plt.plot(range(1, 10), inercia, marker='o')
plt.xlabel("Número de clusters")
plt.ylabel("Inercia")
plt.title("Método del codo")
plt.show()

# 🔹 (Elegimos k = 2 después de ver el codo)
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(df)

# 🔹 Centroides
centroides = kmeans.cluster_centers_

# 🔹 Agregar clusters
df['Cluster'] = kmeans.labels_

# 🔹 Visualización final
plt.scatter(df['Horas de estudio'], df["Calificaciones"], c=df["Cluster"])
plt.scatter(centroides[:, 0], centroides[:, 1],
            marker='x', color='red', s=200, label='Centroides')

plt.xlabel("Horas de estudio")
plt.ylabel("Calificaciones")
plt.title("Clustering con K-Means")
plt.legend()
plt.show()
```
